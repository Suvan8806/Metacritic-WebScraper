```python
#import requests to get the html of the website
import requests

#Use headers as if not metacritic will block webscrapers
url = "https://www.metacritic.com/browse/movie/all/all/2018/metascore/?page=1"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}

data = requests.get(url, headers=headers)
print(data.status_code)  # 200 means success
#print(data.text[:1000])  # Print first 1000 characters to check if content is fetched
```

```python
#Use Beautifulsoup to scrape for the data that is required, check the html id by going to website and inspect element
from bs4 import BeautifulSoup
soup = BeautifulSoup(data.text)
titleData = soup.find_all(class_='c-finderProductCard_title')
descriptionData = soup.find_all('div', class_='c-finderProductCard_description')
dateData = soup.find_all(class_='u-text-uppercase')
ageRatingData = soup.find_all(class_='c-finderProductCard_meta')
scoreData = soup.find_all(class_='c-finderProductCard_score')
```

```python
#Now properly format the data into seperate lists
# For all the Titles
titles = []
for title in titleData:
    # title.text coems in as 1. Roma
    # so split it for the first space only, then get the second element in the list
    titles.append((title.text.split(" ",1))[1])
titles
```

```python
descriptions = []
# For all the Descriptions
for description in descriptionData:
    # description.text simply gives us the description needed
    descriptions.append(description.text)
descriptions
```

```python
dates = []
# For all the dates
for date in dateData:
    # date.text comes in with unnesecary spaces and \n, so simply clean it up by parsing the string
    dates.append((date.text[11:].split("\n"))[0])
dates
```

```python
ageRatings = []
# For all the age Ratings
# Using a counter as ageRating.text gives us a list of items, one with the age rating (even) and one with the metascore(odd), so use a counter and only use the evens
c = 0
for ageRating in ageRatingData:
    if c % 2 == 0:
        # Use a try catch statement as some dont have a age Rating
        try:
          # ageRating.text gives us '\n          Nov 21, 2018\n         \n          \xa0•\xa0\n         Rated R\n        ', so parse the string
          ageRatings.append((ageRating.text.split("Rated", 1)[1]).split("\n")[0][1:])
        except:
          ageRatings.append("NA")
    c += 1
ageRatings
```

```python
scores = []
# For all the scores
for score in scoreData:
    # score.text simply gives us the description needed
    scores.append(score.text)
scores
```

```python
# Create the dataframe to append all the data collected so far
import pandas as pd
data = {
       'Title': titles,
       'Description': descriptions,
       'Date': dates,
       'Age Rating': ageRatings,
       'Score': scores
   }
df = pd.DataFrame(data)
df
```

```python
#Now that I know I can get all the data for any specific page, get it for all the pages by running a for loop for how many pages are available, and changing the url
#get the page numbers

import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}
data = requests.get("https://www.metacritic.com/browse/movie/all/all/2018/metascore/?page=1", headers=headers)
soup = BeautifulSoup(data.text)

pageNumbers = soup.find_all(class_="c-navigationPagination_itemButton")[3].text.split("\n")[1][12:]




url = "https://www.metacritic.com/browse/movie/all/all/2018/metascore/?page=" #without the last digit
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}

#initialize all movies to append df from each page
allMovies = []

for i in range(int(pageNumbers)):

  pageUrl = url + str(i+1) #add the last digit here

  # Get all the data
  data = requests.get(pageUrl, headers=headers)

  soup = BeautifulSoup(data.text)
  titleData = soup.find_all(class_='c-finderProductCard_title')
  descriptionData = soup.find_all('div', class_='c-finderProductCard_description')
  dateData = soup.find_all(class_='u-text-uppercase')
  ageRatingData = soup.find_all(class_='c-finderProductCard_meta')
  scoreData = soup.find_all(class_='c-finderProductCard_score')

  titles = []
  for title in titleData:
      titles.append((title.text.split(" ",1))[1])
  titles;

  descriptions = []
  for description in descriptionData:
      descriptions.append(description.text)
  descriptions;

  dates = []
  for date in dateData:
      dates.append((date.text[11:].split("\n"))[0])
  dates;

  ageRatings = []
  c = 0
  for ageRating in ageRatingData:
      if c % 2 == 0:
          try:
            ageRatings.append((ageRating.text.split("Rated", 1)[1]).split("\n")[0][1:])
          except:
            ageRatings.append("NA")
      c += 1
  ageRatings;

  scores = []
  for score in scoreData:
      scores.append(score.text)
  scores;

  # Create the dataframe to append all the data collected so far
  data = {
        'Title': titles,
        'Description': descriptions,
        'Date': dates,
        'Age Rating': ageRatings,
        'Score': scores
    }
  df = pd.DataFrame(data)
  allMovies.append(df)

allMoviesDf = pd.concat(allMovies, ignore_index=True)
```
